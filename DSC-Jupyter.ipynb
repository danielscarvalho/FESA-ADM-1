```python
# -*- coding: utf-8 -*-
"""
Created on Mon May  9 21:57:23 2022

@author: pro14533
"""
# Pré-processamento, converter CSV com separador , para ; com excel

def fixcomma(num):
    return float(num.replace(",","."))

def fixempty(num):
    if len(num) == 0:
        return 0
    else:
        return int(num)


# encoding >> latin1, ascii ou utf-8
moviedb = open("mymoviedb-v2.csv","r", encoding="latin1")

moviedb.readline(); #Pular uma linha - cabeçalho

mdata=[]

for linha in moviedb:
    l = linha.split(";")

    # 0 Release_Date
    # 1 Title
    # 2 Overview
    # 3 Popularity
    # 4 Vote_Count
    # 5 Vote_Average
    # 6 Original_Language
    # 7 Genre	Poster_Url
    
    if len(l[3]) > 1 and len(l[0].split("/"))==3:
        mdata.append([l[0],fixcomma(l[3]),fixempty(l[4]),fixcomma(l[5])])

print(mdata)
```

```python
import pandas as pd
```

```python
df_mdata = pd.DataFrame (mdata, columns = ['data','vote','popularity','average'])
```
